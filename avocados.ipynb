```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
import datetime
import matplotlib.dates as mdates
```

```python
df = pd.read_csv('avocado 2.csv')
```

```python
df.head()
```

```python
df = df.rename(columns = {'4046' : 'small', '4225':'medium', '4770':'large'})
```

```python
df =(df.drop(['Unnamed: 0'], axis=1))
```

```python
df.Date = pd.to_datetime(df.Date)
```

```python
df.head()
```

```python
df.region.unique()
```

```python
df.describe(include='all')
```

## Visualize average price from 2015-2018 in each region 

Lower prices are are in southern and (Dallas, Houston) and higher prices are in northern and colder areas and in California (Albany, Hartford, Sanfransico) 

```python
plt.figure(figsize=(15,8))
sns.barplot(x='region', y='AveragePrice', data=df)
plt.xticks(rotation=90)
```

### Average price for each year for organic

```python
mask = df.type == 'organic'
sns.catplot(x='AveragePrice', y='region', height=15,data=df[mask], hue='year', kind='point', join=False)
```

### Average price each year for conventional

```python
mask = df.type == 'conventional'
sns.catplot(x='AveragePrice', y='region', height=15,data=df[mask], hue='year', kind='point', join=False)
```

### Average price by month

```python
plt.figure(figsize=(20,12))
sns.lineplot(x='Date', y='AveragePrice', data=df, hue='type')
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.xticks(rotation=45)
```

## Show the distribution of prices organic and conventional

```python
fig = plt.figure(figsize=(15,15))

conv = df.loc[df['type'] == 'conventional']
org = df.loc[df['type'] == 'organic']

sns.distplot(conv.AveragePrice)
sns.distplot(org.AveragePrice)
fig.legend(labels=['conventional', 'organic'])
```

```python
plt.figure(figsize=(15,15))
sns.boxplot(x='AveragePrice', y='type', data=df, palette = 'pastel')
```
